```python
import pandas as pd
from dataset import Spectrum, load_h5
```

```python
data = load_h5("dsetsf.h5")

df = pd.DataFrame(
    columns=[
        "B",
        "theta",
        "phi",
        # "freq",
        # "signal",
    ]
)
for i in data:
    df = pd.concat(
        [
            df,
            pd.DataFrame(
                {
                    "B": i.B,
                    "theta": i.theta,
                    "phi": i.phi,
                    # "freq": list(i.freq),
                    # "signal": list(i.signal),
                },
                index=[0],
            ),
        ],
        ignore_index=True,
    )
```
